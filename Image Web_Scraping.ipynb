### Sarees

```python
# Opening the google images
url = "https://images.google.com/"
driver.get(url)
```

```python
search_bar = driver.find_element_by_xpath('//*[@id="sbtc"]/div/div[2]/input')    # Finding the search bar using it's xpath
search_bar.send_keys("sarees")       # Inputing "sarees" keyword to search images
search_button = driver.find_element_by_xpath('//*[@id="sbtc"]/button')    # Finding the xpath of search button
search_button.click()        # Clicking the search button
```

```python
# 20 time we scroll down by 200 in order to generate more images on the website
for _ in range(20):
    driver.execute_script("window.scrollBy(0,250)")
    
images = driver.find_elements_by_xpath('//img[@class="rg_i Q4LuWd"]')

img_urls = []
img_data = []
for image in images:
    source= image.get_attribute('src')
    if source is not None:
        if(source[0:4] == 'http'):
            img_urls.append(source)
```

```python
for i in range(len(img_urls)):
    if i >= 250:
        break
    print("Downloading {0} of {1} images" .format(i, 250))
    response= requests.get(img_urls[i])
    file = open(r"F:\DATA TRAINED\FLIP ROBO TECH\14th Assignment\saree"+str(i)+".jpg", "wb")
    file.write(response.content)
```
