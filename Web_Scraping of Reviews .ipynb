```python
# opening the link
url = "https://www.flipkart.com/realme-narzo-50a-oxygen-blue-64-gb/product-reviews/itm81d679403c2b4?pid=MOBG6MY4UGGFQWYY&lid=LSTMOBG6MY4UGGFQWYYI8WMJQ&marketplace=FLIPKART"
```

```python
driver.get(url)
```

```python
#creating the empty list
star=[]
full_review=[]
short_review=[]
```

```python
time.sleep(3)
#scrapping the required details
start=0
end=49
for page in range(start,end):#for loop for scrapping 49 page
    mobile_star=driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']/div[1]/div[1]")#scraping brands name by class name='col _2wzgFH K0kLPL'
    for i in mobile_star:
        star.append(i.text)#appending the text in Brand list
    mobile_short_review=driver.find_elements_by_xpath("//p[@class='_2-N8zT']")#scraping description from the xpath
    for i in mobile_short_review:
        short_review.append(i.text)#appending the description in list
    mobile_full_review=driver.find_elements_by_xpath("//div[@class='t-ZTKy']/div/div")# scraping the price from the xpath
    for i in mobile_full_review:
        full_review.append(i.text)
    nxt_button=driver.find_elements_by_xpath("//a[@class='_1LKTO3']")#scraping the list of buttons from the page
    try:
        driver.get(nxt_button[1].get_attribute('href'))#getting the link from the list for next page
    except:
        driver.get(nxt_button[0].get_attribute('href'))
```

```python
#creating a dataframe
df=pd.DataFrame({'mobile_star':star,
                'mobile_full_review':full_review,
                'mobile_short_review':short_review})
#printing dataframe
df
```

```python
df.to_excel('mobile_comments1.xlsx',index=False)
```
