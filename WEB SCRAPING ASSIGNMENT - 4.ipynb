## 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
### Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
#### You need to find following details:
#### A) Rank
#### B) Name
#### C) Artist
#### D) Upload date
#### E) Views

```python
# imports
from selenium import webdriver
import pandas as pd
```

```python
driver = webdriver.Chrome("chromedriver.exe")
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

### Locators

```python
# Locators
rank=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
name=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
artist=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
upload_date=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
views=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
```

```python
Rank = []

for r in rank:
    Rank.append(r.text)
    Rank
```

```python
Rank
```

```python
Name = []
for n in name:
    Name.append(n.text)   
```

```python
Name
```

```python
Artist = []

for a in artist:
    Artist.append(a.text)
```

```python
Artist
```

```python
Upload_Date = []

for u in upload_date:
    Upload_Date.append(u.text)
```

```python
Upload_Date
```

```python
Views = []

for v in views:
    Views.append(v.text)
```

```python
Views
```

```python
# Creating Dataframe
Wikipedia_Youtube = pd.DataFrame({})
Wikipedia_Youtube['rank']=Rank
Wikipedia_Youtube['name']=Name
Wikipedia_Youtube['artist']=Artist
Wikipedia_Youtube['upload_date']=Upload_Date
Wikipedia_Youtube['views']=Views
Wikipedia_Youtube
```

```python
driver.close()
```
